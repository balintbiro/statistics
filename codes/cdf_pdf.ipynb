```python
#importing the required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid
```

```python
#create a function for getting a normal distribution
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
x = np.linspace(1,50,2000)
mean = np.mean(x)
sd = np.std(x)
```

```python
#probability density function
pdf = normal_dist(x,mean,sd)
```

```python
#numerical integration of the pdf results the cumulative distribution function
cdf=cumulative_trapezoid(pdf,x)
```

```python
cdf_values=[]
base=0
for i in pdf:
    base+=i
    cdf_values.append(base/sum(pdf))
```

```python
plt.style.use('seaborn')
fig,ax=plt.subplots(1,2, figsize=(7,3))
ax[0].plot(x,pdf)
ax[1].plot(x[:-1],cdf)
```

```python
plt.style.use('seaborn')
fig,ax=plt.subplots(1,2, figsize=(7,3))
ax[0].plot(x,pdf)
ax[1].plot(x,cdf_values)
```
